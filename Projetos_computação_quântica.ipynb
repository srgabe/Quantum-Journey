# 1) Quantum-Materials Optimizer

Neste projeto, vamos usar o Qiskit (framework da IBM) para resolver um problema de VQE (Variational Quantum Eigensolver).

O VQE é um algoritmo híbrido (quântico-clássico) usado para encontrar o estado de energia fundamental de uma molécula. Então vamos usar computação quântica para simular propriedades de materiais.

### Qubits e Portas Lógicas

Diferente da programação clássica ($0$ ou $1$), na quântica trabalhamos com a Esfera de Bloch. Vamos usar portas como:

-Hadamard (H): Cria sobreposição (o qubit vira $0$ e $1$ ao mesmo tempo).

-CNOT: Cria emaranhamento (o estado de um qubit passa a depender do outro).

Vamos criar um Estado de Bell, que é o exemplo máximo de emaranhamento quântico:

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# 1. Criando um circuito com 2 qubits e 2 bits clássicos
qc = QuantumCircuit(2, 2)

# 2. Aplicando a Porta Hadamard no qubit 0 (Sobreposição)
qc.h(0)

# 3. Aplicando CNOT (Emaranhando o qubit 0 com o 1)
qc.cx(0, 1)

# 4. Medindo os qubits
qc.measure([0, 1], [0, 1])

# 5. Simulando em um computador quântico (AerSimulator)
simulator = AerSimulator()
result = simulator.run(qc).result()
counts = result.get_counts()

print(f"Resultados das medições: {counts}")
# O resultado será aproximadamente 50% '00' e 50% '11'
```

# 2)Simulação da Energia de Ligação da Molécula de Hidrogênio ($H_2$)

Neste projeto, vamos usar o Python para montar um problema de mecânica quântica e resolvê-lo em um simulador quântico.

Trataremos a molécula como um Hamiltoniano ($H$). O computador quântico vai variar os ângulos das portas lógicas (como se fossem botões de ajuste) até encontrar o estado $|\psi\rangle$ que minimiza a energia:

$$E = \frac{\langle\psi|H|\psi\rangle}{\langle\psi|\psi\rangle}$$

```python
pip install qiskit-nature qiskit-algorithms pyscf
```

```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit_nature.second_q.drivers import PySCFDriver
from qiskit_nature.second_q.mappers import JordanWignerMapper
from qiskit_algorithms import VQE
from qiskit_algorithms.optimizers import SLSQP
from qiskit.circuit.library import RealAmplitudes

# A SOLUÇÃO DO IMPORT ESTÁ AQUI:
from qiskit.primitives import StatevectorEstimator

# 1. Preparação das variáveis para o gráfico
distancias = np.arange(0.3, 2.5, 0.2) # De 0.3 a 2.5 Angstroms
energias_vqe = []
energias_classicas = []

print("⚛️ Iniciando simulações quânticas (VQE)...")

for d in distancias:
    # Configuração da molécula para a distância 'd'
    driver = PySCFDriver(atom=f"H 0 0 0; H 0 0 {d}", basis="sto3g")
    problem = driver.run()
    mapper = JordanWignerMapper()
    hamiltonian = mapper.map(problem.second_q_ops()[0])

    # Setup do VQE
    ansatz = RealAmplitudes(num_qubits=hamiltonian.num_qubits, entanglement='full')
    estimator = StatevectorEstimator() # Usando a V2 Primitive
    optimizer = SLSQP(maxiter=50)

    vqe = VQE(estimator, ansatz, optimizer)
    result = vqe.compute_minimum_eigenvalue(hamiltonian)

    energias_vqe.append(result.eigenvalue.real + problem.nuclear_repulsion_energy)
    energias_classicas.append(problem.reference_energy)
    print(f"Distância: {d:.2f} Å | Energia: {energias_vqe[-1]:.4f} Ha")

# 2. Gráfico da Curva de Energia Potencial (Morse Potential)
plt.figure(figsize=(10, 6))
plt.plot(distancias, energias_classicas, 'o-', label='Referência Clássica (Exact)', color='black', alpha=0.3)
plt.plot(distancias, energias_vqe, 's--', label='Simulação Quântica (VQE)', color='blue')

plt.title('Curva de Dissociação da Molécula de $H_2$ via VQE', fontsize=14)
plt.xlabel('Distância Interatômica (Å)', fontsize=12)
plt.ylabel('Energia Total (Hartree)', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

# Modelo de Fermi-Hubbard em Redes

No Grafeno, os elétrons saltam entre os sítios da rede. O Hamiltoniano que descreve isso é:

$$H = -t \sum_{\langle i,j \rangle, \sigma} (c_{i\sigma}^\dagger c_{j\sigma} + h.c.) + U \sum_i n_{i\uparrow} n_{i\downarrow}$$

$t$ (Hopping): A energia de "pulo" entre átomos vizinhos.

$U$ (Interação): A repulsão eletrônica no mesmo sítio (essencial para entender isolantes de Mott ou magnetismo).

```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit_nature.second_q.hamiltonians import FermiHubbardModel
from qiskit_nature.second_q.hamiltonians.lattices import LineLattice, BoundaryCondition
from qiskit_nature.second_q.operators import FermionicOp
from qiskit_nature.second_q.mappers import JordanWignerMapper
from qiskit_algorithms import VQE
from qiskit_algorithms.optimizers import SLSQP
from qiskit.circuit.library import RealAmplitudes
from qiskit.primitives import StatevectorEstimator

# 1. Definindo a rede (2 sítios)
num_sitios = 2
rede = LineLattice(num_nodes=num_sitios, boundary_condition=BoundaryCondition.OPEN)

# Parâmetros Físicos
t = -2.8  # Hopping (eV)
u_valores = np.linspace(0, 15, 8)
energias = []

print("⚛️ Calculando a Transição de Mott (Resolvendo conflito de Spin)...")

for u in u_valores:
    # Construindo o operador fermiônico manualmente para evitar erros de biblioteca
    # Isso garante que o JordanWignerMapper receba exatamente o que ele espera
    label_dict = {}

    # Termos de Hopping (t) - Entre sítios 0 e 1 (Spin Up e Down)
    # No Qiskit Nature, índices 0,1 são Up e 2,3 são Down para 2 sítios
    for i, j in rede.graph.edge_list():
        for s in [0, num_sitios]: # s=0 (Up), s=2 (Down)
            label_dict[f"+_{i+s} -_{j+s}"] = t
            label_dict[f"+_{j+s} -_{i+s}"] = t

    # Termos de Interação (U) - Repulsão no mesmo sítio i
    for i in range(num_sitios):
        # n_up * n_down = (+_i -_i) * (+_{i+num_sitios} -_{i+num_sitios})
        label_dict[f"+_{i} -_{i} +_{i+num_sitios} -_{i+num_sitios}"] = u

    hamiltoniano_fermi = FermionicOp(label_dict, num_spin_orbitals=2*num_sitios)

    # Mapeamento para Qubits (Jordan-Wigner)
    qubit_op = JordanWignerMapper().map(hamiltoniano_fermi)

    # 3. Execução do VQE
    ansatz = RealAmplitudes(num_qubits=qubit_op.num_qubits, entanglement='full')
    vqe = VQE(StatevectorEstimator(), ansatz, SLSQP(maxiter=30))

    res = vqe.compute_minimum_eigenvalue(qubit_op)
    energias.append(res.eigenvalue.real)
    print(f"U = {u:>5.2f} eV | E_ground = {energias[-1]:>8.4f} eV")

# 4. Gráfico Final
plt.figure(figsize=(10, 6))
plt.plot(u_valores, energias, 'o-', color='#C0392B', label='Ground State (Hubbard)')
plt.title('Simulação Quântica: Transição de Mott (2-site Hubbard)', fontsize=14)
plt.xlabel('Interação U (eV)', fontsize=12)
plt.ylabel('Energia (eV)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

Note como a inclinação da curva muda conforme o $U$ aumenta: isso é a competição energética em ação. No início ($U=0$), a energia é puramente cinética (hopping), e conforme você aumenta a repulsão, o sistema "sofre" para acomodar os elétrons, aproximando-se do regime isolante.
